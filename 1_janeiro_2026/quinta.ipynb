```r
# vetor de nomes
nomes <- c(
  "Fernanda",
  "Jaqueline",
  "Dara",
  "Isabela",
  "Jadeir",
  "Gabriella",
  "Valentina",
  "Ana Clara",
  "Hugo Fristo"
)
```

```r
# acessando posições dos vetores
print(nomes[8])
```

```r
# usando índices
nomes[2:4] <- toupper(nomes[2:4])
nomes[c(2,3,4)] <- toupper(nomes[c(2,3,4)])
# usando variável adicional
lista <- c(2,3,4)
nomes[lista] <- toupper(nomes[lista])
```

```r
# vetor de números
idades <- c(19, 25, 24, 24, 25, 26, 22, 20, 46)
```

```r
# cria planilha nomes+idades
dados <- data.frame(
  "nome"=nomes,
  "idade"=idades
)
```

```r
# acessando o data.frame
dados$idade < 22
```

```r
# converte o vetor v/f em posições
which(dados$idade < 22)
```

```r
# usa posições para exibir nomes
# dados[linha,coluna]
dados[which(dados$idade < 22),"nome"]
```

```r
# calcula aposentadoria
dados$falta_para_aposentar <- NA
dados$falta_para_aposentar <- 60 - dados$idade
```

```r
# laço com data.frame
for (linha in 1:nrow(dados)) {
  if (dados$nome[linha] == "Hugo Cristo") {
    dados$nome[linha] <- "Hugo"
  }
}
print(dados$nome)
```

```r
# laço com data.frame
for (linha in 1:nrow(dados)) {
  print(
    paste0(
      "O nome da pessoa é ",
      dados$nome[linha],
      " e a idade dela é ",
      dados$idade[linha],
      "."
    )
  )
}
```

```r
# criando planilha de dados sintéticos
sintetico <- data.frame()

profissoes <- c("Design","Psicologia","Direito","Matemática","Petróleo e Gás","Música")

for (pessoa in 1:10000) {

  # idade aleatória
  idade <- round(runif(1,15,50),0)

  # altura aleatória
  altura <- round(runif(1,1.60,1.90),2)

  # sorteia entre as profissoes
  profissao <- sample(profissoes,1)

  # adiciona linha
  sintetico <- rbind(sintetico, 
                    data.frame(
                      "idade"=idade,
                      "altura"=altura,
                      "profissao"=profissao
                    )
                )
}
```
