```r
# vetor de nomes
nomes <- c(
  "Fernanda",
  "Jaqueline",
  "Dara",
  "Isabela",
  "Jadeir",
  "Gabriella",
  "Valentina",
  "Ana Clara",
  "Hugo Cristo"
)
```

```r
# acessando posições dos vetores
print(nomes[8])
```

```r
# altera apenas
nomes[8] <- toupper(nomes[8])
```

```r
# usando índices
nomes[2:4] <- toupper(nomes[2:4])
nomes[c(2,3,4)] <- toupper(nomes[c(2,3,4)])
# usando variável adicional
lista <- c(2,3,4)
nomes[lista] <- toupper(nomes[lista])
```

```r
# altera todos de uma vez
nomes <- toupper(nomes)
```

```r
# vetor de números
idades <- c(19, 25, 24, 24, 25, 26, 22, 20, 46)
```

```r
# cria planilha nomes+idades
dados <- data.frame(
  "nome"=nomes,
  "idade"=idades
)
```

```r
# acessando o data.frame
dados$idade < 22
```

```r
# converte o vetor v/f em posições
which(dados$idade < 22)
```

```r
# usa posições para exibir nomes
# dados[linha,coluna]
dados[which(dados$idade < 22),"nome"]
```

```r
# calcula aposentadoria
dados$falta_para_aposentar <- NA
dados$falta_para_aposentar <- 60 - dados$idade
```

```r
# remove coluna
dados$nome <- NULL
```
